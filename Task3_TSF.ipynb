# State

```python
# total profits from different states
state_profits = data.groupby('State')['Profit'].sum().sort_values(ascending= False)
```

```python
state_profits.plot.bar(figsize=(18,12))
plt.show()
```

```python
# grouping profits with respect to state and category
state_category_profits = data.groupby(['State','Category'])['Profit'].mean().sort_values(ascending= False)
```

```python
# checking for categories which gave negative profit -- statewise
state_category_profits[state_category_profits<=0].plot.bar(figsize=(18,12))
plt.show()
```

```python
# checking for sub-categories which gave negative profit -- statewise
state_category_subcategory_profits = data.groupby(['State','Category','Sub-Category'])['Profit'].mean().sort_values(ascending= False)

state_category_subcategory_profits[state_category_subcategory_profits<=0].plot.bar(figsize=(18,9))
plt.show()
```

```python
# checking the discounts for categories which gave negative profit -- statewise
state_category_subcategory_discounts = data.groupby(['State','Category','Sub-Category'])['Discount'].mean().sort_values(ascending= False)

state_category_subcategory_discounts[state_category_subcategory_profits<=0].plot.bar(figsize=(18,9))
plt.show()
```

```python
# array containing all the stores ,grouped by sub-category, where negative profit was found .
Shop_list_loss = state_category_subcategory_discounts[state_category_subcategory_profits<=0].index
```

```python
# array containing all the stores ,grouped by sub-category, where profit was found .
shop_list_gain = state_category_subcategory_profits[state_category_subcategory_profits>0].index
```

```python
# array containing top 100 stores ,grouped by sub-category, where profit was found .
shop_list_gain_top100 = state_category_subcategory_profits[state_category_subcategory_profits>0].nlargest(100).index
```

# Conclusion

Shop_list_loss consists all the places, including the sub-category ,where necessary actions must be taken like reducing the discounts and if both sales and profits decrease with this action then try for alternate solutions or items to be sold. We can also not sell these items in these places to reduce loss if nothing works out.

Shop_list_gain contains all the places, including the sub-category, where more plans for increasing sales can be initiated to increase the profits. 
 

Shop_list_gain_top100 consists of top 100 places, including the sub-category, where there is a good demand for the respective items. So, more of these items can be made available at these stores.
